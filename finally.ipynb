```python
import requests
import urllib.request as urllib2
from bs4 import BeautifulSoup
import csv
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
product_list = pd.read_excel("product_list_v03.xlsx", sheet_name="unique_keys")
```

```python
key = []
headers ={"User-Agent": 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'}

for i in product_list["unique_keywords"]:
    key.append(i)
keywords = [y for x in key for y in x.split(',')]
```

```python
nothing = [""]
df = pd.DataFrame({"Date of process":nothing,"Products ":nothing,"Prices ":nothing,"Adress ":nothing,"Keyword ":nothing})
df.to_csv("data.csv",index=False)
print("Csv file for saving data is created!")
```

```python
for key in keywords:
    if ' ' in key:
        key = key.replace(' ', '')
    else:
        key = key
    URL = "https://makromarket.uz/search?term={}".format(key)
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content,'html.parser')
    titles = []
    prices = []
    Urls = []
    product_title = soup.find_all(attrs={"class":"goods-item-name"})
    product_price = soup.find_all(attrs={"class":"card-new-price"})
    product_url = soup.find_all(attrs={"class":"goods-item-img-wrapper"})
    titles = [x.get_text() for x in product_title] #getting all titles in the list
    prices = [y.get_text() for y in product_price] #getting all prices in the list
    Urls = [z.get("href") for z in product_url]#getting all product urls in the list
    
    try:
        with open('data.csv', 'a', newline='', encoding='cp1251') as thething:
            writer = csv.writer(thething)
            for index,title in enumerate(titles):
                writer.writerow([datetime.now(),title,prices[index],URL,key,Urls[index]])
        print("{}'s fineshed processing".format(key)) 
    except UnicodeEncodeError:
        continue
    
    
print(len(keywords)," product informations in Makromarket.uz are exported to data.csv successfully!!!")
    
    
```

```python
for key in keywords:
    if ' ' in key:
        key = key.replace(' ', '')
    else:
        key = key
        
    URL = "http://hm.uz/search/?query={}".format(key)
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content,'html.parser')
    
    title = soup.find_all(attrs={"item__title"})
    price = soup.find_all(attrs={"item__prc-w"})
    url = soup.find_all(attrs={"item__image item__image_gv"})
    titles = [x.get_text() for x in title]
    prices = [y.get_text() for y in price]
    Urls = [z.get('href') for z in url]
    urls = ['http://hm.uz/'+a for a in Urls]
    
    titles = np.array(titles)
    prices = np.array(prices) 
    urls = np.array(urls)
    
    
    try:
        with open('data.csv', 'a', newline='', encoding='cp1251') as thething:
            writer = csv.writer(thething)
            for index,title in enumerate(titles):
                writer.writerow([datetime.now(),title,prices[index],URL,key,urls[index]])
        print("{}'s fineshed processing".format(key)) 
    except UnicodeEncodeError:
        continue
        
print(len(keywords)," product informations in HM.uz are exported to data.csv successfully!!!")
```

```python
for key in keywords:
    if ' ' in key:
        key = key.replace(' ', '')
    else:
        key = key
        
    URL = "https://lebazar.uz/store/korzinka-uz-7/shedevr-78/search/result/products?searchKey={}".format(key)
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content,'html.parser')
    
    title1 = soup.find_all(attrs={"product-title"})
    price1 = soup.find_all(attrs={"product__price clearfix"})

    titles = [x.get_text() for x in title1]
    prices = [y.get_text() for y in price1]
    
    
    try:
        with open('data.csv', 'a', newline='', encoding='cp1251') as thething:
            writer = csv.writer(thething)
            for index,title in enumerate(titles):
                writer.writerow([datetime.now(),title,prices[index],URL,key," "])
        print("{}'s fineshed processing".format(key)) 
    except UnicodeEncodeError:
        continue

print(len(keywords)," product informations in LEBAZAR are exported to data.csv successfully!!!")
```
